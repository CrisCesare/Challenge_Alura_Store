### Importa√ß√£o dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
import matplotlib.ticker as ticker

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja1.head()
```

#1. An√°lise do faturamento








```python
#Somando o faturamento total de cada loja
faturamentos = [
    loja1['Pre√ßo'].sum(),
    loja2['Pre√ßo'].sum(),
    loja3['Pre√ßo'].sum(),
    loja4['Pre√ßo'].sum()]

#V√°riaveis para apresenta√ß√£o anual
dfs = [loja1, loja2, loja3, loja4]
faturamentoanual = []

#Soma de cada loja anualmente
for i in dfs:
  i['Data da Compra'] = pd.to_datetime(i['Data da Compra'], dayfirst=True)
  i['Ano da Compra'] = i['Data da Compra'].dt.year
  faturamentoanual.append(i.groupby('Ano da Compra')['Pre√ßo'].sum())

#V√°riaveis para apresenta√ß√£o trimestral
faturamento_trimestral = []

#Soma de cada loja trimestralmente
for i in dfs:
    i['Ano_Trimestre'] = i['Data da Compra'].dt.to_period('Q').astype(str)
    i['Ano_Trimestre'] = i['Ano_Trimestre'].apply(lambda x:x.replace('Q', 'T'))
    faturamento_trimestral.append(i.groupby('Ano_Trimestre')['Pre√ßo'].sum())

#Apresenta√ß√£o do resultado
for i in range(len(faturamentos)):
  print(f'O faturamento total da Loja {i+1} √© de R$ {faturamentos[i]:,.2f}')
```

```python
#V√°riaveis para apresenta√ß√£o gr√°fica
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
anos = list(faturamentoanual[0].index)
valores_por_loja = [list(faturamento.values) for faturamento in faturamentoanual]
trimestres_todos = sorted(set().union(*[loja.index for loja in faturamento_trimestral]))
valores_por_loja = [list(faturamento.values) for faturamento in faturamento_trimestral]

#Defini√ß√£o de cores e tamanhos dos gr√°ficos
cores_linhas = ['#3344F5', '#9966FF', '#96D7FF', '#FFC0CB']
fig, ax = plt.subplots(1, 2, figsize=(20, 5))
fig, ay = plt.subplots(figsize=(17, 5))

#Gr√°fico de barras - Faturamento Total
bars = ax[0].bar(lojas, faturamentos, color='#1A237E')
ax[0].set_title('Faturamento Total')
ax[0].yaxis.set_major_formatter(ticker.StrMethodFormatter('R$ {x:,.2f}'))
ax[0].set_ylim(1350000, 1550000)

media_faturamento = sum(faturamentos) / len(faturamentos)
ax[0].axhline(media_faturamento, color='green', linestyle='--', linewidth=2, label=f'M√©dia: R$ {media_faturamento:,.2f}')
ax[0].legend()

for bar in bars:
    height = bar.get_height()
    ax[0].text(bar.get_x() + bar.get_width() / 2, height + 3500, f'R$ {height:,.2f}', ha='center', va='bottom', color='black', fontsize=9)

#Gr√°fico de linhas - Anual
for i, loja in enumerate(faturamentoanual):
  anos = loja.index
  valores = loja.values
  ax[1].plot(anos, valores, marker='o', label=f'Loja {i + 1}', color = cores_linhas[i])
  for ano, valor in zip(anos, valores):
    ax[1].text(ano, valor, f'{valor:,.0f}', color = cores_linhas[i], ha='center', va='bottom', fontsize=9)

anos_todos = sorted(set().union(*[loja.index for loja in faturamentoanual]))
ax[1].margins(0.055)
ax[1].set_xticks(anos_todos)
ax[1].set_xticklabels(anos_todos)
ax[1].set_title('Faturamento por Ano')
ax[1].legend(title='Lojas')
ax[1].yaxis.set_major_formatter(ticker.StrMethodFormatter('R$ {x:,.2f}'))

#Gr√°fico de linhas - Trimestral
for i, loja in enumerate(faturamento_trimestral):
    trimestres = loja.index.astype(str)
    valores = loja.values
    ay.plot(trimestres, valores, marker='o', label=f'Loja {i + 1}', color=cores_linhas[i])
    for trimestre, valor in zip(trimestres, valores):
        ay.text(trimestre, valor, f'{valor:,.0f}', color=cores_linhas[i], ha='center', va='bottom', fontsize=9)

ay.margins(0.055)
ay.set_xticks(range(len(trimestres_todos)))
ay.set_xticklabels([str(trimestre).replace('T', ' T') for trimestre in trimestres_todos], rotation=45)
ay.set_title('Faturamento por Trimestre')
ay.legend(title='Lojas')
ay.yaxis.set_major_formatter(ticker.StrMethodFormatter('R$ {x:,.2f}'))

#Ajustes finais dos gr√°ficos
plt.subplots_adjust(wspace=6)
plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria


```python
#Quantidade de vendas por categoria
venda_categorias = [
    loja1['Categoria do Produto'].value_counts().head(5),
    loja2['Categoria do Produto'].value_counts().head(5),
    loja3['Categoria do Produto'].value_counts().head(5),
    loja4['Categoria do Produto'].value_counts().head(5)]

#Apresenta√ß√£o do Top 5 categorias
for i in range(len(venda_categorias)):
  print(f'As categorias mais populares (Top 5) da Loja {i+1} s√£o: \n{venda_categorias[i]}')
  if i > 2:
    continue
  print()
```

```python
# Cores dos gr√°ficos
cores = ['#3344F5', '#9966FF', '#96D7FF', '#FFC0CB', '#FF6961', '#77DD77', '#FFD700', '#00BFFF', '#FF4500']
categorias_unicas = set()

# Definindo 4 gr√°ficos um ao lado do outro
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
fig.suptitle('Top 5 - Vendas por Categoria', fontsize=28)

#Gerando dados dos gr√°ficos
for i in range(4):
    categorias_unicas.update(venda_categorias[i].index)
    cores_map = {categoria: cores[j % len(cores)] for j, categoria in enumerate(sorted(categorias_unicas))}
    total = venda_categorias[i].sum()
    axes[i].pie(venda_categorias[i], labels=venda_categorias[i].index, colors=[cores_map[categoria] for categoria in venda_categorias[i].index],
                autopct=lambda pct: f'{pct:.1f}%\n({int(round(pct * total / 100))})', startangle=90, counterclock=False)
    axes[i].set_title(lojas[i])

# Ajustes finais dos gr√°ficos
plt.subplots_adjust(wspace=5)
plt.tight_layout()
plt.show()
```

# 3. M√©dia de Avalia√ß√£o das Lojas

```python
#M√©dia da avalia√ß√£o por loja
avaliacoes = [
    round(loja1['Avalia√ß√£o da compra'].mean(),2),
    round(loja2['Avalia√ß√£o da compra'].mean(),2),
    round(loja3['Avalia√ß√£o da compra'].mean(),2),
    round(loja4['Avalia√ß√£o da compra'].mean(),2)]

#V√°riavel para m√©dia anualmente
avaliacaoanual = []

#M√©dia de cada loja anualmente
for i in dfs:
  agrupamento = i.groupby('Ano da Compra')['Avalia√ß√£o da compra'].mean()
  avaliacaoanual.append(agrupamento.round(2))

#Quantidade de avalia√ß√µes por nota
avaliacao_notas = [
    loja1['Avalia√ß√£o da compra'].value_counts().sort_index().head(),
    loja2['Avalia√ß√£o da compra'].value_counts().sort_index().head(),
    loja3['Avalia√ß√£o da compra'].value_counts().sort_index().head(),
    loja4['Avalia√ß√£o da compra'].value_counts().sort_index().head()
]

#Apresenta√ß√£o do resultado
for i in range(len(avaliacoes)):
  print(f'A m√©dia de avalia√ß√£o da Loja {i+1} √© de {avaliacoes[i]}')
```

```python
#Defini√ß√£o da distribui√ß√£o dos gr√°ficos
fig, ax = plt.subplots(1, 3, figsize=(30, 7))

#Gr√°fico 1
bars = ax[0].barh(lojas, avaliacoes, color='#1A237E')
ax[0].set_title('M√©dia de Avalia√ß√£o dos Clientes', fontsize=16)
ax[0].set_xlim(3.50, 4.10)
ax[0].invert_yaxis()

for bar in bars:
    width = bar.get_width()
    ax[0].text(width, bar.get_y() + bar.get_height() / 2, f'{width: .2f}', ha='left', va='center', color='black', fontsize=10)

#Gr√°fico 2
for i, loja in enumerate(avaliacaoanual):
  anos = loja.index
  valores = loja.values
  ax[1].plot(anos, valores, marker='o', label=f'Loja {i + 1}', color = cores_linhas[i])
  for ano, valor in zip(anos, valores):
    ax[1].text(ano, valor, f'{valor:,.2f}', color = cores_linhas[i], ha='center', va='bottom', fontsize=9)

anos_todos = sorted(set().union(*[loja.index for loja in avaliacaoanual]))
ax[1].margins(0.055)
ax[1].set_xticks(anos_todos)
ax[1].set_xticklabels(anos_todos)
ax[1].set_title('Avalia√ß√£o por Ano', fontsize=16)
ax[1].legend(title='Lojas')

#Gr√°fico 3
notas = sorted(set().union(*[avaliacao_notas[i].index for i in range(4)]))
notas_com_texto = [f'Nota {nota}' for nota in notas]  # Adicionando 'Nota' antes de cada valor de nota
largura = 0.2

for i in range(4):
    valores = [avaliacao_notas[i].get(nota, 0) for nota in notas]
    posicoes = [j + largura * i for j in range(len(notas))]
    ax[2].bar(posicoes, valores, largura, label=lojas[i], color=cores_linhas[i])
    for pos, valor in zip(posicoes, valores):
        ax[2].text(pos, valor + 0.5, f'{valor}', ha='center', va='bottom', fontsize=10)

posicoes_centrais = [j + largura * 1.5 for j in range(len(notas))]  # Definindo posi√ß√µes centrais
ax[2].set_xticks(posicoes_centrais)
ax[2].set_xticklabels(notas_com_texto)  # Aplicando as etiquetas com texto adicional
ax[2].legend(title='Lojas')
ax[2].set_title('Quantidade Avalia√ß√µes por Nota', fontsize=16)

#Ajustes finais dos gr√°ficos
plt.subplots_adjust(wspace=5)
plt.tight_layout()
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
#Quantidade de produtos vendidos
venda_produtos = [
    loja1['Produto'].value_counts(),
    loja2['Produto'].value_counts(),
    loja3['Produto'].value_counts(),
    loja4['Produto'].value_counts()]

#Variav√©is para lista de produtos mais e menos vendidos
produtos_mais_vendidos = []
produtos_menos_vendidos = []

#Extra√ß√£o dos produtos mais e menos vendidos por loja
for vendas in venda_produtos:
    max_vendas = vendas.max()
    min_vendas = vendas.min()
    produtos_max = vendas[vendas == max_vendas]
    produtos_min = vendas[vendas == min_vendas]
    produtos_mais_vendidos.append(produtos_max)
    produtos_menos_vendidos.append(produtos_min)

#Apresenta√ß√£o do resultado
for i in range(len(produtos_mais_vendidos)):
    print(f"Loja {i + 1} - Produtos mais vendidos:")
    for produto, quantidade in produtos_mais_vendidos[i].items():
        print(f" - {produto}: {quantidade} vendas")
    print()

    print(f"Loja {i + 1} - Produtos menos vendidos:")
    for produto, quantidade in produtos_menos_vendidos[i].items():
        print(f" - {produto}: {quantidade} vendas")
    print()
    if i > 2:
        continue
    print("-" * 40)
    print()
```

```python
#Vari√°vel para agrupar itens
produtos_agrupados = []

#Agrupando os dados
for i in range(len(venda_produtos)):
    loja_atual = f"Loja {i + 1}"
    produtos_da_loja = produtos_mais_vendidos[i].index.tolist() + produtos_menos_vendidos[i].index.tolist()

    for produto in produtos_da_loja:
        dados_produto = {
            "Produto": produto,
            "Loja de Origem": loja_atual,
            "Vendas": {}
        }
        for j in range(len(venda_produtos)):
            loja_analisada = f"Loja {j + 1}"
            quantidade_vendida = venda_produtos[j].get(produto, 0)  # Obter quantidade ou 0 caso n√£o exista
            dados_produto["Vendas"][loja_analisada] = quantidade_vendida

        produtos_agrupados.append(dados_produto)

#Convertendo os dados em DataFrame
dados_para_grafico = []
for produto_info in produtos_agrupados:
    for loja, quantidade in produto_info["Vendas"].items():
        dados_para_grafico.append({
            "Produto": produto_info["Produto"],
            "Loja": loja,
            "Quantidade Vendida": quantidade
        })

df_grafico = pd.DataFrame(dados_para_grafico)

#Configura√ß√£o do gr√°fico
plt.figure(figsize=(28, 8))
produtos_unicos = df_grafico["Produto"].unique()
lojas_unicas = df_grafico["Loja"].unique()
largura = 0.8 / len(lojas_unicas)

for i, loja in enumerate(lojas_unicas):
    df_loja = df_grafico[df_grafico["Loja"] == loja]
    x = range(len(produtos_unicos))
    y = [df_loja[df_loja["Produto"] == produto]["Quantidade Vendida"].sum() for produto in produtos_unicos]

    barras = []
    for pos, produto in enumerate(produtos_unicos):
      if produto in produtos_mais_vendidos[i]:
        hatch = '////'
        facecolor = cores_linhas[i]
        edgecolor = 'white'
      elif produto in produtos_menos_vendidos[i]:
        hatch = '...'
        facecolor = cores_linhas[i]
        edgecolor = 'white'
      else:
            hatch = None
            facecolor = cores_linhas[i]
            edgecolor = 'none'

      barra = plt.bar(pos + i * largura, y[pos], width=largura, label=loja if pos == 0 else "", facecolor=facecolor, edgecolor=edgecolor, hatch=hatch)
      barras.append(barra)

    for barra in barras:
        for b in barra:
            plt.text(b.get_x() + b.get_width() / 2, b.get_height() + 0.5, int(b.get_height()), ha="center", va="bottom", fontsize=12)

handles = [Patch(facecolor=cores_linhas[i], label=loja) for i, loja in enumerate(lojas_unicas)]
plt.legend(handles=handles, title="Lojas", loc="upper right")

#Ajustes finais do gr√°fico
plt.xticks([pos + largura * (len(lojas_unicas) - 1) / 2 for pos in range(len(produtos_unicos))], produtos_unicos, fontsize=12)
plt.title("Desempenho de Produtos por Loja", fontsize=16)
plt.tight_layout()
plt.show()
```

# 5. Frete M√©dio por Loja


```python
#M√©dia do frete por loja
valor_frete = [
    round(loja1['Frete'].mean(),2),
    round(loja2['Frete'].mean(),2),
    round(loja3['Frete'].mean(),2),
    round(loja4['Frete'].mean(),2)]

#V√°riavel para m√©dia anualmente
freteanual = []

#M√©dia de cada loja anualmente
for i in dfs:
  agrupamento = i.groupby('Ano da Compra')['Frete'].mean()
  freteanual.append(agrupamento.round(2))

media_frete = round((loja1['Frete'].sum() + loja2['Frete'].sum() + loja3['Frete'].sum() + loja4['Frete'].sum()) / (len(loja1['Frete']) + len(loja2['Frete']) + len(loja3['Frete']) + len(loja4['Frete'])),2)

#Apresenta√ß√£o do resultado
for i in range(len(valor_frete)):
  print(f'A m√©dia de frete da Loja {i+1} √© de R$ {valor_frete[i]}')
```
